#### Define global parameters to run L63 model:

```python
sigma = 10.0       # Parameter sigma
rho   = 28.0       # Parameter rho
beta  = 8.0 / 3.0  # Parameter beta
```

#### Define the L63 model:

```python
def lorenz63(state, t):
        """
        Calculates the derivatives for the Lorenz63 system.
        
        Parameters:
        state: vector [x, y, z] representing the current state.
        t: current time (not used in Lorenz63 but included for compatibility).
        
        Returns:
        A vector of derivatives [dx/dt, dy/dt, dz/dt].
        """
        x, y, z = state  # Unpack the state vector
        dx = sigma * (y - x)              # Equation for dx/dt
        dy = x * (rho - z) - y            # Equation for dy/dt
        dz = x * y - beta * z             # Equation for dz/dt
        return np.array([dx, dy, dz])     # Return the derivative vector
```

#### Run Nature L63 model:

```python
name = 'lorenz63_run'

t0 = 0
#t1 = 410
t1 = 80000
#t1 = 80
dt = 0.01
dt_output = 0.01
tvec = np.arange(t0, t1, dt)
t = np.arange(t0,t1-10, dt_output)
print("time vector: ", t)

# Define initial conditions and integration parameters:
np.random.seed(2)
#initial_state = np.random.normal(0, np.sqrt(2), 3)
initial_state = [ 8.20747939, 10.0860429,  23.86324441]


# Create state_vector object:
sv = state_vector(initial_state, t=np.arange(t0,t1-10, dt_output), name=name)
#print(sv) # print the state vector object to see the attributes and methods


# Define parameters for the Lorenz63 system:
h             = 0.01                         # Integration step size
t_final       = t1                        # Total integration time
#t_final       = 410.0                        # Total integration time


# Create the RK4 solver:
solver = run_RK4(lorenz63, sigma, beta, rho, initial_state, h, t_final)

# Run the solver:
trajectory = solver.integrate()
#trajectory = trajectory[1000:, :]
#tvec       = tvec[1000:]
print("Trajectory: ", trajectory)

# Set the trajectory attribute of the state_vector object:
sv.setTrajectory(trajectory)

# Store the nature run:
outfile = name+'.pkl'
sv.save(outfile)

# Plot the results:
solver.plot_trajectory()
solver.plot_3d_trajectory()
```
